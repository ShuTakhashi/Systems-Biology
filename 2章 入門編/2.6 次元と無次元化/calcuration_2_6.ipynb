## 2.6 次元と無次元化



```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=0.0
dt=0.01
endTime=15.0
steps=endTime/dt
```

```python
a = np.array([[1],[4],[2],[8]])
c = np.array([[1],[2],[0.5],[2]])

print(a)
print(c)
```

```python
x_0 = np.zeros((len(a),1))
x=x_0

print(x)
```

```python
t_array = np.arange(0,endTime,dt)
x_array = np.array([[],[],[],[]])
```

```python
for i in range(int(steps)):
    t+= dt
    x+= dt*(a-c*x)
    x_array = np.append(x_array, x, axis=1)
```

```python
plt.plot(t_array, x_array[0,:],color='red')
plt.plot(t_array, x_array[1,:],color='blue')
plt.plot(t_array, x_array[2,:],color='green')
plt.plot(t_array, x_array[3,:],color='yellow')


plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```

```python
x_scaled = x_array/(a/c)

print(x_array/(a/c))
```

```python
plt.plot(t_array, x_scaled[0,:],color='red')
plt.plot(t_array, x_scaled[1,:],color='blue')
plt.plot(t_array, x_scaled[2,:],color='green')
plt.plot(t_array, x_scaled[3,:],color='yellow')

plt.xlabel('t')
plt.ylabel('X(t)')
plt.show()
```

```python
t_constant = t_array*c

print(t_constant)
```

```python
plt.plot(t_constant[0,:], x_scaled[0,:],color='red')
plt.plot(t_constant[1,:], x_scaled[1,:],color='blue')
plt.plot(t_constant[2,:], x_scaled[2,:],color='green')
plt.plot(t_constant[3,:], x_scaled[3,:],color='yellow')

plt.xlabel('τ')
plt.ylabel('X(τ)')
plt.show()
```
